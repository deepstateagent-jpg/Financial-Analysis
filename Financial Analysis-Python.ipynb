```python
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import pandas_datareader as web
import yfinance as yf
import talib as ta
```

```python
Apple_stock_price = yf.download(tickers="AAPL")
```

```python
Apple_stock_price
```

```python

sma_window = 50
Apple_stock_price['SMA_' + str(sma_window)] = ta.SMA(Apple_stock_price['Close'], sma_window)
```

```python
Apple_stock_price
```

```python
plt.plot(Apple_stock_price['Close'],label='Closing Price')
```

```python
plt.show()
```

```python
plt.plot(Apple_stock_price['Close'], label='Closing Price')
plt.plot(Apple_stock_price['SMA_' + str(sma_window)], label='SMA ' + str(sma_window))
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Apple Inc. Stock Price with Simple Moving Average')
plt.legend()
```

```python
plt.show()
```

```python
ema_window = 100
Apple_stock_price['EMA_' + str(ema_window)] = ta.EMA(Apple_stock_price['Close'], ema_window)
plt.plot(Apple_stock_price['Close'], label='Closing Price')
plt.plot(Apple_stock_price['EMA_' + str(ema_window)], label='EMA_' + str(ema_window))
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Apple Inc. Stock Price with Exponential Moving Average')
plt.legend()
```

```python
plt.show()
```

```python
rsi_window = 40
Apple_stock_price['RSI_' + str(rsi_window)] = ta.RSI(Apple_stock_price['Close'], rsi_window)
plt.plot(Apple_stock_price['Close'], label='Closing Price')
plt.plot(Apple_stock_price['RSI_' + str(rsi_window)], label='RSI_' + str(rsi_window))
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Apple Inc. Stock Price with Relative Strength Index')
plt.legend()
plt.show()
```

```python
linear_regression = ta.LINEARREG(Apple_stock_price['Close'], 200)
plt.plot(linear_regression, label='Linear Regression')
```

```python
# Calculate MACD
#macd = ta.trend.MACD(Apple_stock_price['Close'], window_slow=26, window_fast=12, window_signal=9)
#apple_stock_price['macd'] = macd.macd()
#apple_stock_price['signal'] = macd.macd_signal()
# Calculate MACD
macd, signal, _ = ta.MACD(Apple_stock_price['Close'], fastperiod=12, slowperiod=26, signalperiod=9)
Apple_stock_price['macd'] = macd
Apple_stock_price['signal'] = signal
```

```python
# Plot MACD
plt.figure(figsize=(10, 6))
plt.plot(Apple_stock_price.index, Apple_stock_price['macd'], label='MACD')
plt.plot(Apple_stock_price.index, Apple_stock_price['signal'], label='Signal Line')
plt.bar(Apple_stock_price.index, Apple_stock_price['macd'] - Apple_stock_price['signal'], label='MACD - Signal', color='gray')
plt.xlabel('Date')
plt.ylabel('MACD')
plt.title('MACD for Apple Stock')
plt.legend()
plt.grid(True)
plt.show()
```
